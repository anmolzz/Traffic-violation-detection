```python
import requests
import cv2
from google.colab.patches import cv2_imshow

def detect_helmet_violation(image_path):
    """
    Detect helmet violations in an image using Roboflow API.

    Args:
        image_path (str): Path to the local image file.

    Returns:
        tuple: Annotated image and a list of bounding box regions for violations.
    """
    # API details for helmet detection
    API_KEY = "tN3bnCvsIbF4A4WbOovO"
    PROJECT_ID = "bike-helmet-detection-2vdjo"
    MODEL_VERSION = "2"

    # Step 1: Send image to Roboflow API for inference
    with open(image_path, "rb") as image_file:
        response = requests.post(
            f"https://detect.roboflow.com/{PROJECT_ID}/{MODEL_VERSION}?api_key={API_KEY}",
            files={"file": image_file}
        )

    # Parse JSON response
    predictions = response.json().get("predictions", [])

    # Load the image using OpenCV
    image = cv2.imread(image_path)

    violation_detected = False
    violation_regions = []  # To store regions of interest for license plates

    for prediction in predictions:
        # Extract bounding box coordinates and class label
        x_center, y_center = prediction["x"], prediction["y"]
        width, height = prediction["width"], prediction["height"]
        class_label = prediction["class"]

        # Calculate bounding box corners
        x0 = int(x_center - width / 2)
        y0 = int(y_center - height / 2)
        x1 = int(x_center + width / 2)
        y1 = int(y_center + height / 2)

        # Draw bounding box on the image
        color = (0, 255, 0) if class_label == "With Helmet" else (0, 0, 255)  # Green for "With Helmet", Red for "Without Helmet"
        cv2.rectangle(image, (x0, y0), (x1, y1), color, 2)

        # Annotate class label on the bounding box
        cv2.putText(image, class_label, (x0, y0 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale=0.6, color=color, thickness=2)

        # Check for violations
        if class_label == "Without Helmet":
            violation_detected = True

            # Save region of interest for license plates (assuming it's below the rider)
            violation_regions.append((x0, y1 + 10, x1, y1 + int(height)))

    # Display results in Colab
    print("Helmet Violation Detected:" if violation_detected else "No Helmet Violation Detected")
    cv2_imshow(image)  # Display annotated image inline in Colab

    # Save annotated image locally (optional)
    cv2.imwrite("helmet_violation_output.jpg", image)

    return image, violation_regions
```

```python
from PIL import Image
import pytesseract
import cv2

def detect_license_plate(image_path):
    """
    Detect and extract text from license plates in an image.
    """
    # Roboflow API details
    API_KEY = "tN3bnCvsIbF4A4WbOovO"
    PROJECT_ID = "indian-car-bike-number-plate"
    MODEL_VERSION = "2"

    # Send image to Roboflow API for inference
    with open(image_path, "rb") as image_file:
        response = requests.post(
            f"https://detect.roboflow.com/{PROJECT_ID}/{MODEL_VERSION}?api_key={API_KEY}",
            files={"file": image_file}
        )

    predictions = response.json().get("predictions", [])

    # Load image using OpenCV
    image = cv2.imread(image_path)

    extracted_texts = []

    for prediction in predictions:
        # Extract bounding box coordinates
        x_center, y_center = prediction["x"], prediction["y"]
        width, height = prediction["width"], prediction["height"]
        x0 = int(x_center - width / 2)
        y0 = int(y_center - height / 2)
        x1 = int(x_center + width / 2)
        y1 = int(y_center + height / 2)

        # Crop and preprocess license plate region
        cropped_image = image[y0:y1, x0:x1]
        processed_plate = preprocess_license_plate(cropped_image)

        # Perform OCR on processed region using Tesseract
        custom_config = r'--oem 3 --psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
        license_plate_text = pytesseract.image_to_string(processed_plate, config=custom_config).strip()

        extracted_texts.append(license_plate_text)

        # Annotate detected text on original image for debugging
        cv2.rectangle(image, (x0, y0), (x1, y1), (0, 255, 0), 2)
        cv2.putText(image, license_plate_text, (x0, y0 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale=0.6, color=(255, 255, 255), thickness=2)

    # Display annotated image in Colab
    cv2_imshow(image)

    return extracted_texts
```

```python
import os
import requests

def send_violation_data(api_url, licenseplate, violationtype, image_path):
    if not os.path.exists(image_path):
        return {"error": f"File not found: {image_path}"}

    try:
        with open(image_path, "rb") as img_file:
            data = {"licensePlate": licenseplate, "violationType": violationtype}
            files = {"violationImage": img_file}


            print("Sending Data:", data)
            print(" Sending File:", image_path)

            response = requests.post(api_url, data=data, files=files)
            print("Status Code:", response.status_code)
            print("Response Text:", response.text)

            response.raise_for_status()
            return response.json()

    except requests.exceptions.RequestException as e:
        return {"error": f"API request failed: {e}"}
```

```python
def preprocess_license_plate(cropped_image):
    gray = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)  # Grayscale
    blurred = cv2.bilateralFilter(gray, 11, 17, 17)         # Noise reduction
    thresh = cv2.adaptiveThreshold(
        blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
    )
    return thresh
```
